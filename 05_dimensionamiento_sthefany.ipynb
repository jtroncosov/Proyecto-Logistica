# 5. Dimensionamiento

## 1. Título

Dimensionamiento del centro logístico.

## 2. Objetivo de la sección

Determinar la superficie necesaria del centro logístico.

## 3. Contexto teórico breve

El dimensionamiento permite calcular el espacio necesario para almacenar productos según la demanda.

## 4. Metodología / enfoque

Se usa la fórmula:

Superficie = Demanda × Espacio × Factor de seguridad

```python
demanda_mensual = 10000
espacio_por_unidad = 0.5
factor_seguridad = 1.2

superficie = demanda_mensual * espacio_por_unidad * factor_seguridad

print(f"Superficie requerida: {superficie} m2")
```

## 5. Desarrollo

Se calcula la superficie considerando demanda, espacio por unidad y un factor de seguridad.

## 6. Resultados

La superficie requerida es de 6.000 m².

## 7. Conclusión de la sección

El centro logístico necesita aproximadamente 6.000 m² para operar correctamente.
