# 2. Introducción

## 1. Título

Introducción al problema logístico de distribución de insumos médicos en Providencia.

## 2. Objetivo de la sección

Definir el contexto del problema logístico, justificar la comuna de estudio y caracterizar el tipo de producto a almacenar.

## 3. Contexto teórico breve

La logística de distribución busca organizar el flujo de productos desde un centro de almacenamiento hacia los puntos de demanda, reduciendo costos, tiempos de entrega y recorridos innecesarios.

En zonas urbanas con alta concentración de demanda, como Providencia, la ubicación del centro logístico es clave para mejorar la eficiencia del sistema.

## 4. Metodología / enfoque

Se analiza la comuna de Providencia como zona de demanda, considerando su concentración urbana, conectividad vial y presencia de centros de salud.

Además, se define el producto a almacenar como insumos médicos, debido a que requieren entregas frecuentes, ordenadas y oportunas.

## 5. Desarrollo

El crecimiento urbano y la alta concentración de demanda en Providencia generan desafíos en la distribución eficiente de productos.

Se selecciona Providencia debido a:

- Alta densidad de demanda.
- Buena conectividad vial.
- Cercanía con comunas estratégicas como Recoleta, Santiago y Ñuñoa.
- Presencia de uso de suelo compatible.

El producto a almacenar corresponde a insumos médicos.

Características:

- Tipo: insumos médicos.
- Condiciones: almacenamiento limpio y ordenado.
- Volumen: medio.
- Rotación: alta.
- Impacto logístico: requiere entregas frecuentes y rápidas.

## 6. Resultados

Como resultado, se define que el estudio se enfoca en la distribución urbana de insumos médicos hacia Providencia, considerando la localización de un centro logístico como una decisión estratégica para reducir tiempos y mejorar el servicio.
